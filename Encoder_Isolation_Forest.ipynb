```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data",names=["symboling", "normalized_losses",
"make", "fuel_type", "aspiration","num_doors", "body_style", "drive_wheels",
"engine_location", "wheel_base", "length", "width", "height", "curb_weight",
"engine_type", "num_cylinders", "engine_size", "fuel_system", "bore", "stroke",
"compression_ratio", "horsepower", "peak_rpm", "city_mpg", "highway_mpg", "price"],na_values='?')
df
```

```python
df.dropna(subset=['price'],inplace=True)
df.reset_index(drop=True,inplace=True)
df
```

```python
df.dtypes
```

```python
df.dtypes.index
```

```python
keys=list(df.dtypes.index[df.dtypes!=object])
for key in keys:
    df[key].fillna(df[key].mean(),inplace=True)
df
```

```python
keys_category=list(df.dtypes.index[df.dtypes==object])
for key in keys_category:
    df[key].fillna(df[key].mode()[0],inplace=True)
df
```

```python
from sklearn.ensemble import IsolationForest
iso_forest=IsolationForest()
```

```python
iso_forest.fit(df[['length']])
predicted=iso_forest.predict(df[['length']])
ind_of_outliers=np.where(predicted==-1)
ind_of_outliers
```

```python
outliers=df['length'].iloc[ind_of_outliers]
outliers
```

```python
df['num_doors'].unique()
```

```python
df['num_cylinders'].unique()
```

```python
dictionary={'two':2,'three':3,'four':4,'six':6,'five':5,'twelve':12,'eight':8}
```

```python
for key in dictionary.keys():
    df[['num_doors','num_cylinders']]=df[['num_doors','num_cylinders']].replace(key,dictionary[key])
```

```python
df
```

```python
dummies=pd.get_dummies(df[["body_style", "drive_wheels"]])
df=pd.concat((df,dummies),axis=1)
df.drop(columns=["body_style", "drive_wheels"],inplace=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder
label_enc=LabelEncoder()
features=['make', 'aspiration', 'engine_location','fuel_type']
for feature in features:
    df[feature]=label_enc.fit_transform(df[feature])
df
```

```python
arr = []
for val in df['fuel_system']:
    if 'pfi' in val:
        arr.append(1)
    else:
        arr.append(0)
df['fuel_system'] = arr
df['fuel_system']
```

```python
arr1=[]
for val in df['engine_type']:
    if 'ohc' in val:
        arr1.append(1)
    else:
        arr1.append(0)
df['engine_type'] = arr1
df['engine_type']
```

```python
input_features=df.iloc[:,:-1]
output_features=df.iloc[:,-1]
input_features
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
scaled_input_features=pd.DataFrame(ss.fit_transform(input_features),columns=input_features.keys())
scaled_input_features
```
