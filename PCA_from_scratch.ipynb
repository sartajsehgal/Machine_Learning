```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_digits
digits=load_digits()
```

```python
type(digits)
```

```python
X=digits.data
print(X)
```

```python
X.shape
```

```python
X=pd.DataFrame(X)
```

```python
print(X)
```

```python
Y=digits.target
Y=pd.DataFrame(Y)
print(Y)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_scaled=pd.DataFrame(ss.fit_transform(X))
print(X_scaled)
```

```python
Cov_matrix=X_scaled.cov()
print(Cov_matrix)
```

```python
eigen_values, eig_vectors=np.linalg.eig(Cov_matrix)
```

```python
eigen_values
```

```python
eig_pairs=[]
for i in range(len(eigen_values)):
  eig_pairs.append((eigen_values[i],eig_vectors[:,i]))
```

```python
eig_pairs
```

```python
def first_ele(tup):
  return tup[0]

sorted_eig_pairs=sorted(eig_pairs, reverse=True, key=first_ele)
```

```python
for i in sorted_eig_pairs:
  print(i[0])
```

```python
result = pd.DataFrame()
#1st 40 features are selected 
for i in range(40):
    PC = X_scaled.dot(sorted_eig_pairs[i][1].T)
    result['PC'+str(i+1)] = PC
result['number'] = Y
result
```
